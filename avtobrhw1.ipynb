```python
from natasha import (
    Segmenter,
    
    NewsEmbedding,
    NewsMorphTagger,
    NewsSyntaxParser,
    NewsNERTagger,
    
    Doc
)

from natasha import MorphVocab
```

```python
def preproc(text):
    segmenter = Segmenter()

    emb = NewsEmbedding()
    morph_tagger = NewsMorphTagger(emb)
    syntax_parser = NewsSyntaxParser(emb)

    doc = Doc(text)

    doc.segment(segmenter)
    doc.tag_morph(morph_tagger)
    doc.parse_syntax(syntax_parser)

    ner_tagger = NewsNERTagger(emb)
    doc.tag_ner(ner_tagger)

    morph_vocab = MorphVocab()

    for token in doc.tokens:
        token.lemmatize(morph_vocab)

    for span in doc.spans:
        span.normalize(morph_vocab)

    # 
    
    dictposition = dict()
    for el in doc.tokens:
        try:
            start = el.start
        except:
            start  = 0
        
        dictposition[start] = el.lemma

    # замена нескольких токенов нормализованным словосочетанием
    for el in doc.spans:
        try:
            start = el.start
        except:
            start  = 0
        # удаление токенов входящих в словосочетание из словаря
        for tok in el.tokens:
            try:
                tokstart = tok.start
            except:
                tokstart  = 0
            del dictposition[tokstart]
        # add phrase
        dictposition[start] = el.normal

    sd = sorted(dictposition.items())
    lem_text_tokenized = [el[1] for el in sd] # if el[1] not in punctuation
    lem_text = ' '.join(lem_text_tokenized)
    return lem_text
```

разметка наташей частей речи в автоматически выделенных кс

```python
def get_pos(text):
    segmenter = Segmenter()
    emb = NewsEmbedding()
    morph_tagger = NewsMorphTagger(emb)
    doc = Doc(text)
    doc.segment(segmenter)
    doc.tag_morph(morph_tagger)
    res = []
    for token in doc.tokens:
        if token.pos == 'PROPN' or token.pos == 'X':
            res.append('NOUN')
        else:
            res.append(token.pos)
    return ' + '.join(res)
```

фильтрация по синтаксическим шаблонам

```python
dict_with_synt = {'rake':[], 
                   'textrank':[], 
                   'tfidf':[],
                   }

for key in diff_kw_results:
    print(key)
    for kwlist in diff_kw_results[key]:
        res_for_text = []
        for kw in kwlist:
            pos_str = get_pos(kw)
            # print(pos_str)
            if pos_str in synt:
                res_for_text.append(kw)
        dict_with_synt[key].append(res_for_text)
```

```python
# lemmatize gold
gold_lem = []
for kwlist in gold:
    textres = []
    for kw in kwlist:
        segmenter = Segmenter()
        emb = NewsEmbedding()
        morph_tagger = NewsMorphTagger(emb)
        doc = Doc(kw)
        doc.segment(segmenter)
        doc.tag_morph(morph_tagger)
        morph_vocab = MorphVocab()
        lem_kw = []
        for token in doc.tokens:
            token.lemmatize(morph_vocab)
            lem_kw.append(token.lemma)
        textres.append(' '.join(lem_kw))
    gold_lem.append(textres)

gold_lem
```
